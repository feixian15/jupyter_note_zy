### 以四皇后为例说明：

下面这个结果是错误的，因为考虑到各个皇后的顺序问题，第一个皇后可以放到任意一行（只需要放到第0行），所以出现了许多重复的结果，相当于对每一个结果做了一次全排列

```python
all_list = []  ## 主要是解决点的重复问题，八皇后问题点是没有顺序的
def solve(n, point_list):
    if n==0 : 
        print(point_list)
        print_list= []
        for i in point_list:
            print_list.append(i[1])
        all_list.append(point_list)
    else:
        for i in range(4):
            for j in range(4):
                temp = True
                for k in point_list:
                    if i==k[0] or j==k[1] or i+j==k[0]+k[1] or j-i==k[1]-k[0]:
                        temp = False
                        break
                #if (i,j) in all_list: temp = False
                if temp: 
                    point_list.append((i,j))
                    solve(n-1,point_list)
                    point_list.pop()
```
